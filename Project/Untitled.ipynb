## Feature Engineering

```python
from haversine import haversine
```

```python
df_['booking_created'] = pd.to_datetime(df_['booking_created'])
df_['day_of_week'] = df_['booking_created'].dt.dayofweek
df_['hour_of_day'] = df_['booking_created'].dt.hour
```

```python
df_['mobile_site_booking'].value_counts()
```

```python
df_['distance_km'] = df_.apply(lambda row: haversine((row['from_lat'], row['from_long']), (row['to_lat'], row['to_long'])), axis=1)
```
